```python
# Install required libraries
!pip install pandas
!pip install numpy

# Import libraries
import pandas as pd
import numpy as np

# Load the dataset
df = pd.read_csv("https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DS0321EN-SkillsNetwork/datasets/dataset_part_1.csv")
df.head(10)

#Data Analysis
#Missing Values Analysis

# Calculate percentage of missing values
df.isnull().sum()/len(df)*100

#Data Types
# Check data types
df.dtypes
```

```python
#Task 1: Launches per Site

# Calculate number of launches per site
launch_site_counts = df['LaunchSite'].value_counts()
print(launch_site_counts)
```

```python
#Task 2: Orbit Occurrences

# Calculate number of occurrences for each orbit type
orbit_counts = df['Orbit'].value_counts()
print(orbit_counts)
```

```python
#Task 3: Mission Outcomes

# Calculate landing outcomes
landing_outcomes = df['Outcome'].value_counts()
print(landing_outcomes)

# Print outcomes with indices
for i, outcome in enumerate(landing_outcomes.keys()):
    print(i, outcome)

# Define unsuccessful outcomes
bad_outcomes = set(landing_outcomes.keys()[[1, 3, 5, 6, 7]])
print(bad_outcomes)
```

```python
#Task 4: Create Landing Class Labels
# Create binary classification labels
landing_class = []
for outcome in df['Outcome']:
    if outcome in bad_outcomes:
        landing_class.append(0)
    else:
        landing_class.append(1)

# Add to dataframe
df['Class'] = landing_class
df[['Class']].head(8)

# Calculate success rate
success_rate = df["Class"].mean()
print(f"Success rate: {success_rate:.2%}")

# Preview final dataframe
df.head(5)
```

```python
# Save to CSV for next steps
df.to_csv("dataset_part_2.csv", index=False)
```
